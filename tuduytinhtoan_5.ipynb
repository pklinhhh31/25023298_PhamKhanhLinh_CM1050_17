#W5A1 max of two

```python
def max_of_two( x, y ):
    if x > y:
        return x
    return y
print(max_of_two(3, 5))
```

W5A3 kiểm tra số nguyên tố

```python
def prime(n):
  for i in range(2, int(n**0.5)+1):
    if n%i == 0:
      return False
  return True
print(prime(12))
```

#W5A4 số hoàn hảo

```python
def perfect_num(num):
  sum=0
  for i in range(1, num//2 + 1):
    if num % i == 0:
      sum += i
  if num == sum: return True
  return False
print(perfect_num(28))
```

#W5A5

```python
def test(lst, k):
  for i in range(len(lst)):
    if lst[i]==k:
      return i
  return -1
print(test([1, 2, 3,5], 9))

```

#W5A6 factorial

```python
def factorial(n):
  result=1
  for i in range(2,n+1):
    result *= i
  return result
print(factorial(5))
```

#W5A7 máy tính bỏ túi

```python
num1 = float(input('nhập vào số đầu: '))
operat = input('nhập toán tử: ')
num2 = float(input('nhập số thứ 2: '))
result=0
if operat == '+':
  result = num1+num2
elif operat == '-':
  result = num1 - num2
elif operat == '*':
  result = num1* num2
elif operat == '/':
  result= num1 / num2
print(f'kết quả là: {round(result,2)}')
```

#W5A9 tổng các số trong số n

```python
def sum(n):
  num = list(n)
  tong = 0
  for i in num:
    tong += int(i)
  return tong
n= input('nhập số n ')
print(f'tổng các số của n là {sum(n)}')
```

```python
def sum(n):
  tong = 0
  while n >= 1:
    tong += n %10
    n = n//10
  return tong
n = int(input('nhập số n '))
print(f'tổng các số của n là {sum(n)}')
```

#W5A10 hai từ đẳng cấu

```python

def dangcau(word1, word2):
  if len(word1) != len(word2): return False
  dic= {}
  for i in range(len(word1)):
    if dic.get(word2[i]) == None: #kiểm tra xem chữ i có ảnh chưa
      dic[word2[i]] = word1[i]    # chưa có thì thêm vào từ điển
    elif dic[word2[i]] != word1[i]: #kiểm tra hạt nhân có giống nhau ko
      return False
  return True
word1= list(input('nhập từ thứ nhất '))
word2 = list(input('nhập từ thứ hai '))
print(dangcau(word1, word2))
```
