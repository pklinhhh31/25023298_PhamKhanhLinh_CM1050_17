```python
toan= float(input('nhập điểm toán: '))
ly= float(input('nhập điểm lý: '))
hoa= float(input('nhập điểm hóa: '))
anh= float(input('nhập điểm anh: '))
diem_tb= (toan+ly+hoa+anh)/4
if toan < 4 or hoa < 4 or ly <4 or anh < 4:
  print('học lực kém')
else:
  if diem_tb >= 9:
      print('học lực xuất sắc')
  elif diem_tb >= 8.5:
      print('học lực giỏi')
  elif diem_tb >= 7:
      print('học lực khá')
  elif diem_tb >= 5.5:
      print('học lực trung bình')
  else:
      print('học lực yếu')
```

```python
chieu_cao= float(input('nhập chiều cao(m): '))
can_nang= float(input('nhập cân nặng: '))
BMI= can_nang/(chieu_cao**2)
phan_loai= ""
if BMI < 18.5:
  phan_loai= 'Thiếu cân'
elif 18.5 <= BMI < 25:
  phan_loai= 'Bình thường'
elif 25 <= BMI < 30:
  phan_loai= 'Thừa cân'
else:
  phan_loai= 'Béo phì'
print(f"BMI của bạn là: {BMI:.2f}")
print(f"phân loại: {phan_loai}")
```
