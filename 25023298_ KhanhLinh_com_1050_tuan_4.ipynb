##W3A16 Rounding

```python
a= float(input('nhập số làm tròn:'))
down = int(a)     # làm tròn xuống
up = down + 1    # ~~~~~~~~ lên
if a - down == 0:
  up = down
if a - down >= 0.5:
  near = down + 1
else:
  near = down

print(up, down, near)
```

##W3A17 Cấp số nhân

```python
a, b, c, d = map(float, input('nhập 4 số(có cách): ').split())
if a != 0:
  q = b/a
  if b*q == c and c*q == d:
    print(a, b, c, d, 'là cấp số nhân')
  else: print(a, b, c, d, 'không phải cấp số nhân')
elif a == b == c == d == 0:
  print(a, b, c, d, 'là cấp số nhân')
else: print(a, b, c, d, 'không phải cấp số nhân')
```
